```python
import sys
import ply.lex as lex

class AgroLexer(object):
    tokens = (
        'SEMILLA', 'MES', 'VARIABLE', 'OPERATOR', 'STRING', 'INT', 'FLOAT',
        'IF', 'ELSE', 'LPAREN', 'RPAREN', 'LBRACE', 'RBRACE', 'SEMI', 'EQUALS',
        'SEMBRAR', 'EN', 'FERTILIZAR', 'CON', 'FUMIGAR', 'POR', 'ALERTAHELADA', 'APLICAR', 'RIEGO', 'LITROS'
    )

    t_LPAREN   = r'\('
    t_RPAREN   = r'\)'
    t_LBRACE   = r'\{'
    t_RBRACE   = r'\}'
    t_SEMI     = r';'
    t_EQUALS   = r'=='
    t_OPERATOR = r'(<=|>=|<|>)'

    reserved = {
        'if': 'IF', 'else': 'ELSE', 'Sembrar': 'SEMBRAR', 'en': 'EN',
        'Fertilizar': 'FERTILIZAR', 'con': 'CON', 'Fumigar': 'FUMIGAR',
        'por': 'POR', 'AlertaHelada': 'ALERTAHELADA', 'aplicar': 'APLICAR',
        'Riego': 'RIEGO', 'litros': 'LITROS',
        '"Papa"': 'SEMILLA', '"Maiz"': 'SEMILLA',
        'Temperatura': 'VARIABLE', 'Humedad': 'VARIABLE', 'Lluvia': 'VARIABLE',
        'Enero': 'MES', 'Febrero': 'MES', 'Marzo': 'MES', 'Abril': 'MES',
        'Mayo': 'MES', 'Junio': 'MES', 'Julio': 'MES', 'Agosto': 'MES',
        'Septiembre': 'MES', 'Octubre': 'MES', 'Noviembre': 'MES', 'Diciembre': 'MES'
    }

    def t_STRING(self, t):
        r'"[^"\n]*"'
        t.type = self.reserved.get(t.value, 'STRING')
        return t

    def t_ID(self, t):
        r'[a-zA-Z_][a-zA-Z0-9_]*'
        if t.value in self.reserved:
            t.type = self.reserved[t.value]
            return t
        return None

    def t_FLOAT(self, t):
        r'\d+\.\d+'
        t.value = float(t.value)
        return t

    def t_INT(self, t):
        r'\d+'
        t.value = int(t.value)
        return t

    t_ignore = ' \t'

    def t_newline(self, t):
        r'\n+'
        t.lexer.lineno += len(t.value)

    def t_error(self, t):
        print(f"LÉXICO: Carácter ilegal '{t.value}' en línea {t.lexer.lineno}")
        t.lexer.skip(1)

    def __init__(self):
        # 💡 TRUCO PARA PYTHON 3.14 EN NOTEBOOKS:
        # Inyectamos una ruta de archivo falsa en el módulo __main__ 
        # para evitar el error interno de PLY al inicializar.
        main_module = sys.modules['__main__']
        if not hasattr(main_module, '__file__'):
            main_module.__file__ = 'notebook_fake_file.py'
            
        self.lexer = lex.lex(module=self, optimize=False)

# Instanciamos los elementos para que queden disponibles
lexer_instance = AgroLexer()
lexer = lexer_instance.lexer
tokens = lexer_instance.tokens
```
